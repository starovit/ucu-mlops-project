```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/dataset.csv', index_col=0)
df['Date'] = pd.to_datetime(df['Date']).apply(lambda x: x.date())
```

```python
df.groupby('Commodity').count().sort_values('Average', ascending = False)[:20]
```

```python
# Filter the data for 'Ginger'
ginger_data = df[df['Commodity'] == 'Ginger']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(ginger_data['Date'], ginger_data['Average'], label='Ginger Average Price')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Average Price of Ginger Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
